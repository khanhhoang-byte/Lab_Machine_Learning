```python
import networkx as nx
import matplotlib.pyplot as plt
import random

graph =  nx.read_gml("sample_graph.gml")

# В файл уже записаны позиции для отрисовки (граф был сгенерирован networkx)
pos = nx.get_node_attributes(graph, "pos")

# назначаем каждому узлу случайный цвет. В дальнейшем цвет нужно назначить исходя из номера сообщества

# отрисуем граф
node_to_color = [random.randint(0,10) for _ in graph.nodes]
plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(graph, pos, alpha=0.4)
nx.draw_networkx_nodes(
    graph,
    pos,
    nodelist=graph.nodes,
    node_size=80,
    node_color=node_to_color,
    cmap=plt.cm.tab20,
)
```
