```python
import numpy as np
from matplotlib import pyplot as plt


x = np.arange(-5,6) 
f = x - (x^3)/6 + (x^5)/120 - (x^7)/5040
plt.title("Matplotlib 1") 
plt.xlabel("x axis caption") 
plt.ylabel("f axis caption") 
plt.plot(x,f) 
plt.show()
```

```python
g = np.sin(x)
plt.title("Matplotlib 2") 
plt.xlabel("x axis caption") 
plt.ylabel("g axis caption") 
plt.plot(x,g) 
plt.show()
```
