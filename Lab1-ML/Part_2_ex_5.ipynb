```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

```python
x = np.linspace(0,6,100)
f = lambda x: np.sin(x)
y = f(x)
dydx = derivative(f,x)

plt.figure(figsize=(12,5))
plt.plot(x,y,label='y=f(x)')
plt.plot(x,dydx,label="y=f'(x)")
plt.legend()
plt.grid(True)

plt.show()
```
