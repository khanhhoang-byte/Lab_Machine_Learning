```python
#Выведите график числа заказов за каждый прошедший месяц.
order_df = pd.read_csv('orders.csv')
order_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

Date, Quantity = order_df['OrderDate'], order_df['Quantity']
plt.bar(Date,Quantity)
plt.show()
```

```python
#Определите самые дорогие заказы (сгруппировав по OrderID)
order_df["OrderSum"] = order_df["UnitPrice"]*order_df["Quantity"] + (order_df["UnitPrice"]*order_df["Quantity"]*order_df["Discount"])
```

```python
order_df
```

```python
order_df.groupby(['OrderID'])['OrderSum'].max()
```

```python
#Введите столбец "Profit" - разность между доходами и приблизительной закупочной стоимостью проданных товаров.
order_df
```
