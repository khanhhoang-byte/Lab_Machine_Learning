```python
def get_dummies_for_test(train_columns, test_df, dummy_sep):
    initial_df = pd.get_dummies(test_df, prefix_sep=dummy_sep)
#     missing_in_train = set(test_df.columns) - set(train_columns)
    missing_in_test = set(train_columns) - set(initial_df.columns)
    
    for c in missing_in_test:
        initial_df[c] = 0
    return initial_df.loc[:,train_columns].copy()
```

```python
housing_test_dummies = get_dummies_for_test(housing_train_dummies.columns, housing_test, DUMMY_SEP)
housing_val_dummies = get_dummies_for_test(housing_train_dummies.columns, housing_val, DUMMY_SEP)
```
